```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(font_scale = 2)
```

```python
movies = pd.read_csv('Final.csv')
```

## Revenue per genres

```python
genres= pd.read_csv('genres_final.csv')
```

```python
a = genres.columns.tolist()
```

```python
a = ['Crime',
 'Drama',
 'Comedy',
 'Action',
 'Thriller',
 'Adventure',
 'Science Fiction',
 'Animation',
 'Family',
 'Romance',
 'Mystery',
 'Music',
 'Horror',
 'Fantasy',
 'Documentary',
 'War',
 'Western',
 'History',
 'Foreign',
 'TV Movie']
```

```python
lst = []
for i in a: 
    movies[f'revenue_{i}'] = movies[i]*movies['revenue']
    lst.append(f'revenue_{i}')
```

```python
lst
```

```python
movies[lst].mean()
```

```python
import numpy as np
movies[lst]=np.where(movies[lst]==0,np.nan,movies[lst])
```

```python
movies[lst].mean().sort_values(ascending=False).plot.bar(figsize=(12,6))
#,logy=True)
plt.ylabel('revenue',fontsize=25)
plt.title('Revenue by Genres',fontsize=25)
```

# Revenue and countries

```python
countries = pd.read_csv('Countries.csv')
```

```python
countries = pd.DataFrame(countries.name.value_counts())
```

```python
countries.reset_index(inplace=True)
```

```python
countries.rename(columns={'index':'Name','name':'value'}, inplace=True)
```

```python
countries
```

```python
plt.figure(figsize=(20,20))
plt.xticks(rotation=90)
plt.xlabel('Name', fontsize=20)
plt.ylabel('Value', fontsize=20)
sns.barplot(x="Name", y="value", data=countries.head(30))
```
