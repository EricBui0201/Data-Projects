# 2. Modelling:

## 2.1 Fitting:

```python
import statsmodels.api as sm
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
d_YN={'Yes':1,'No':0}
df[['International plan','Voice mail plan']] = df[['International plan','Voice mail plan']].apply(lambda x: x.map(d_YN))
```

```python
df_d =pd.get_dummies(data=df, columns=['State'], drop_first =True)
```

```python
df_d
```

```python
df_d.drop('Total_charge', axis=1, inplace=True)
```

```python
from statsmodels.tools.tools import add_constant

y= df_d.Churn
X = df_d.drop('Churn', axis=1)
X= add_constant(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=18)
```

```python
model = sm.Logit(y_train, X_train)
```

```python
result = model.fit()
```

```python
result.summary()
```

## 2.2 Model Evaluation:

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = round(result.predict(X_train))
```

```python
confusion_matrix(y_train, y_pred)
```

```python
y_train.value_counts()/y_train.shape[0]
```

```python
y_pred.value_counts()/y_pred.shape[0]  # Data is highly imblanced!!!!
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_sk = LogisticRegression(random_state=18, class_weight = 'balanced').fit(X,y)
```

```python
y_pred_sk = model_sk.predict(X_test)
```

```python
confusion_matrix(y_pred_sk,y_test)
```

```python
y_pred_sm= round(result.predict(X_test))
```

```python
confusion_matrix(y_pred_sm, y_test)
```
