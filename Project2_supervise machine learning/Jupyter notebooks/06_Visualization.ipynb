## Revenue per genres

```python
genres= pd.read_csv('genres_final.csv')
```

```python
genres.drop(['Unnamed: 0', 'Unnamed: 0.1', 'id', 'movie_id', 'name'], axis=1, inplace=True)
```

```python
a = genres.columns.tolist()
```

```python
a = ['Crime',
 'Drama',
 'Comedy',
 'Action',
 'Thriller',
 'Adventure',
 'Science Fiction',
 'Animation',
 'Family',
 'Romance',
 'Mystery',
 'Music',
 'Horror',
 'Fantasy',
 'Documentary',
 'War',
 'Western',
 'History',
 'Foreign',
 'TV Movie']
```

```python
lst = []
for i in a: 
    movies[f'revenue_{i}'] = movies[i]*movies['revenue']
    lst.append(f'revenue_{i}')

```

```python
lst
```

```python
movies[lst].mean()
```

```python
import numpy as np
movies[lst]=np.where(movies[lst]==0,np.nan,movies[lst])
```

```python
movies[lst].mean().sort_values(ascending=False).plot.bar(figsize=(12,6))
#,logy=True)
plt.ylabel('revenue',fontsize=15)
plt.title('Revenue by Genres',fontsize=25)
```
