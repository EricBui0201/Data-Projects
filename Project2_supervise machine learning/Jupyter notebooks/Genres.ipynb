```python
import pandas as pd
import numpy as np
```

```python
genres = pd.read_csv('Genres.csv')
```

```python
genres.head(5)
```

```python
genres.drop(['Unnamed: 0','id'], axis=1, inplace=True)
```

```python
genres.sort_values('movie_id', inplace=True)
```

```python
genres.head(5)
```

```python
list_of_genres=genres.name.unique().tolist()
```

```python
list_of_genres
```

```python
for i in list_of_genres:
    genres[i] = 0
```

```python
genres
```

```python
for i in list_of_genres:
    genres[i]= genres.name.str.contains(i).astype(int)
```

```python
genres = genres.groupby('movie_id').agg('sum')
```

```python
genres.reset_index(inplace=True)
```

```python
genres.to_csv('genres_final.csv')  # save for future usage
```
