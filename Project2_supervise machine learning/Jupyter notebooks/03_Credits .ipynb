```python
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
```

```python
credits= pd.read_csv('/Users/linhdao/Data-Projects/Project2_supervise machine learning/Data Prep/credits.csv')
credits = credits.rename(columns={"id": "movie_id"}) 
#Renamed beacause we have id in side the crew and the cast columns



# Getting the CAST:
cast_1 = []
for i,r in enumerate(credits.cast): 
    j = eval(r)
    for k in j: 
        k['movie_id'] = credits['movie_id'][i]
    cast_1.append(j)

cast=[]
for i in cast_1:
    cast.extend(i)  #use extend instead of append because we want only 1 list, not 1ists inside the lis

cast = json_normalize(cast)

```

# GROUPING CASTS:

```python
cast
```

```python
cast = cast[cast['order']<2] #Get the 2 main casts only
```

```python
cast_name = pd.DataFrame(cast.name.value_counts())
# Create a dataframe to group the actors according to the popularity
```

```python
cast_name.describe()
```

```python
# 75% of the actors only appeared maximum 2 times as main casts, statistically means that only 25% of the data points
# actually make sense. However, as we think cast is an important factor effecting Revenue, we keep it. 
```

```python
cast_name.drop(cast_name[cast_name['name'] < 3].index, inplace = True)
```

```python
group_labels = ['1', '2', '3']
```

```python
bins = pd.cut(cast_name['name'],[2,40,70,104], labels=group_labels)
```

```python
bins
```

```python
cast_name['Group']=bins
```

```python
cast_name.reset_index(inplace=True)
```

```python
cast_name.rename(columns={'index':'name','name':'count'}, inplace=True)
```

```python
cast_gr = pd.merge(cast_name,cast,on='name',how='inner') 
# Merging with cast data frame, the table now will have additaional column named 'Group'
```

```python
cast_gr = cast_gr[['movie_id','Group', 'order']]
```

```python
# creating table with movie_id as index
cast1=cast_gr.pivot_table(index='movie_id', columns='order', values='Group',aggfunc='last')
```

```python
cast1.reset_index(inplace=True)
cast1
```

```python
cast1.rename(columns={0:'Cast 1',1:'Cast 2'},inplace=True)
cast1["movie_id"]=cast1["movie_id"].astype(str)
```

```python
cast1.to_csv('Cast_final.csv', index=False)
```
